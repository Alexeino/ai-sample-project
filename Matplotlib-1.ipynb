## Matplotlib example workflow

```python
# 0. Improt matplotlib
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup lot
fig, ax = plt.subplots(figsize=(10,10))

# 3. Plot the data
ax.plot(x,y)

# 4. Customize Plot
ax.set(title="Simple Plot",xlabel="x-axis",ylabel="y-axis")

# 5. Save & Show 
fig.savefig("numpy-images/sample-plot.png")
```

## Making figures with numpy arrays

```python
import numpy as np
# Create some data
x = np.linspace(0,10,100)
x
```

```python
# Plot the data
fig, ax = plt.subplots()
ax.plot(x, x**2)
```

```python
# Scatter Plot
fig, ax = plt.subplots()
ax.scatter(x,np.exp(x))
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x,np.sin(x))
```

```python
# Make a plot from dicitonary
prices = {"Almond butter":10,"Peanut butter":8,"Cashew butter":12}
fig,ax = plt.subplots()
ax.bar(prices.keys(),prices.values())
ax.set(title="Manish's nut butter store",ylabel="Prices ($)");
```

```python
fig, ax = plt.subplots()
ax.barh(prices.keys(),prices.values()); # Horizontal bar
```

```python
# Make some data for histogram and plot it
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x)
```

## Two options for subplots

```python
# Subplot option 1
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax1.plot(x,x/2)
ax2.scatter(np.random.random(10),np.random.random(10))
ax3.bar(prices.keys(),prices.values())
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
# Plot to each different index
ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[1,0].bar(prices.keys(),prices.values())
ax[1,1].hist(np.random.randn(1000));
```

## Plotting form pandas Dataframe

```python
# Make a dataframe
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
ts = pd.Series(np.random.randn(1000),index=pd.date_range("1/1/2020",periods=1000))
ts.plot()
```

```python
ts = ts.cumsum()
ts.plot()
```

```python
car_sales
```

```python
car_sales["Price"] = (
    car_sales["Price"]
    .str.replace(r'[$,]', '', regex=True)
    .astype(float)
    .astype(int)
)
car_sales
```

```python
car_sales["Sales Date"] = pd.date_range("1/1/2020",periods=len(car_sales))
car_sales
```

```python
car_sales["Total Sales"] = car_sales["Price"].cumsum()
car_sales
```

```python
# Let's plot the total sales
car_sales.plot(x="Sales Date",y="Total Sales")
```

```python
car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter")
```

```python
car_sales.plot(x="Make",y="Odometer (KM)")
```

```python
car_sales
```

```python
make_odometer = car_sales.groupby("Make",as_index=False).agg({"Make":"first","Odometer (KM)":"sum"})
make_odometer
```

```python
make_odometer
```

```python
make_odometer.plot(x="Make",y="Odometer (KM)",kind="bar")
```
