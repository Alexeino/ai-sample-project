## Dot product 

```python
np.random.seed(0)
mat1 = np.random.randint(0,10,size=(5,3))
mat2 = np.random.randint(0,10,size=(5,3))
mat1, mat2
```

```python
mat1 * mat2 # Element-wise multiplication
```

```python
# Dot Product
np.dot(mat1,mat2)
```

The inside number of the dimensions must match meaning m X n and n X o are only allowed to be used in dot product that's why a 5X3 and 5X3 matrix can't be dot product.

```python
np.dot(mat1, mat2.T)
```

```python
np.random.seed(0)
nut_butter_sales = np.random.randint(20,size=(5,3))
nut_butter_sales
```

```python
nut_butter_df = pd.DataFrame(nut_butter_sales,index=["Mon","Tue","Wed","Thurs","Fri"],columns=["Almond butter","Peanut butter","Cashew butter"])
nut_butter_df
```

```python
prices = np.array([10,8,12])
prices
```

```python
prices.shape
```

```python
prices = prices.reshape((1,3))
prices.shape
```

```python
butter_prices_df = pd.DataFrame(prices,index=["Price"],columns=["Almond butter","Peanut butter","Cashew butter"])
butter_prices_df
```

```python
prices, nut_butter_sales
```

```python
nut_butter_sales.shape, prices.shape
```

```python
reshaped_prices = prices.reshape(3,1)
reshaped_prices.shape
```

```python
# total_sales = np.dot(prices,nut_butter_sales)
total_sales = np.dot(nut_butter_sales, reshaped_prices)
total_sales
```

```python
nut_butter_df
```

```python
butter_prices_df
```

```python
nut_butter_df["Weekly sales"] = total_sales
nut_butter_df
```
